```python
import pandas as pd
import csv

df_school = pd.read_csv('./Resources/schools_complete.csv')
df_students = pd.read_csv('./Resources/students_complete.csv')


df_students
```

```python
df_school 
```

```python
N_Schools = df_school['School ID'].shape
N_Schools
```

```python
df_school['size'] = df_school['size'].astype(int)
N_Students = df_school['size'].sum()
N_Students
```

```python
TOTAL_BUDGET = df_school['budget'].sum()
TOTAL_BUDGET
```

```python
Average_math = df_students['math_score'].mean()
Average_math
```

```python
Average_reading = df_students['reading_score'].mean()
Average_reading
```

```python
Pass_math = df_students[df_students['math_score']>=70]
Pass_math_students = Pass_math['Student ID']
P_math_int = len(Pass_math_students)

P_math_por = (P_math_int * 100 ) / N_Students
P_math_por
```

```python
Pass_reading = df_students[df_students['reading_score']>=70]
Pass_reading_students = Pass_reading['Student ID']
P_reading_int = len(Pass_reading_students)

P_reading_por = (P_reading_int * 100 ) / N_Students
P_reading_por
```

```python
Pass_R_M = df_students[(df_students['math_score']>=70) & (df_students['reading_score'] >=70)]
Pass_R_M_stu = Pass_R_M['Student ID'] 
Pass_R_M_INT = len(Pass_R_M_stu)
P_r_m_por = (Pass_R_M_INT * 100 ) / N_Students
P_r_m_por 
```

```python
Schools_math_score = df_students.groupby("school_name")[["math_score"]].mean()
Schools_math_score
```

```python
Schools_read_score = df_students.groupby("school_name")[["reading_score"]].mean()
```

```python
new1 = pd.merge(df_Schoool_Summary, Schools_math_score, left_on='School name', right_on='school_name')
```

```python
new2 = pd.merge(new1, Schools_read_score, left_on='School name', right_on='school_name')
```

```python
totales = df_students["school_name"].value_counts()
totales = totales.to_frame()
totales
```

```python
dic_schools = {}
averages = []
school_names = []
for index, row in df_students.iterrows():
    
    dic_schools[row['school_name']] = [0,0,0]

for index, row in df_students.iterrows():
    
    if int(row['math_score']) >= 70:
        dic_schools[row['school_name']][0] = dic_schools[row['school_name']][0] + 1
    if int(row['reading_score']) >= 70:
        dic_schools[row['school_name']][1] = dic_schools[row['school_name']][1] + 1
    if int(row['math_score']) >= 70 and int(row['reading_score']) >= 70:
        dic_schools[row['school_name']][2] = dic_schools[row['school_name']][2] + 1

for index, row in totales.iterrows():
    
    dic_schools[index][0] = (dic_schools[index][0]*100) / int(row['school_name'])
    dic_schools[index][1] = (dic_schools[index][1]*100) / int(row['school_name'])
    dic_schools[index][2] = (dic_schools[index][2]*100) / int(row['school_name'])

for value in dic_schools:
    
    averages.append(dic_schools[value])
    school_names.append(value)

print(averages)
print(school_names)
```

```python
df = pd.DataFrame(averages,
                  columns=['% pass math','% pass reading','% overall passing'],
                  index=school_names)
df = df.reset_index()
df = df.rename(columns={'index':'school_name'})
df
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
Spending_Ranges = []
```

```python
for index, row in df_Schoool_Summary_new.iterrows():
    
    if int(row['Per student budget']) <= spending_bins[1]:
        Spending_Ranges.append(labels[0])
    if int(row['Per student budget']) >= spending_bins[1] and int(row['Per student budget']) <= spending_bins[2]:
        Spending_Ranges.append(labels[1])
    if int(row['Per student budget']) >= spending_bins[2] and int(row['Per student budget']) <= spending_bins[3]:
        Spending_Ranges.append(labels[2])
    if int(row['Per student budget']) >= spending_bins[3] and int(row['Per student budget']) <= spending_bins[4]:
        Spending_Ranges.append(labels[3])

Spending_Ranges
```
